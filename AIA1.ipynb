```python
# Load the LLaMA 3 model with quantization
model_name = "meta-llama/Meta-Llama-3-8B-Instruct"
device = "cuda"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    quantization_config=BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_quant_type="nf4",
    ),
)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# File paths for storing user profile and chat history
USER_PROFILE_FILE = "user_profile.json"
CHAT_HISTORY_FILE = "chat_history.json"
```

```python

# Load or initialize user profile
def load_user_profile():
    try:
        with open(USER_PROFILE_FILE, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        return {}

def save_user_profile(profile):
    with open(USER_PROFILE_FILE, "w") as f:
        json.dump(profile, f)

# Load or initialize chat history
def load_chat_history():
    try:
        with open(CHAT_HISTORY_FILE, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        return []

def save_chat_history(history):
    with open(CHAT_HISTORY_FILE, "w") as f:
        json.dump(history, f)

# Function to generate responses using LLaMA 3
def generate_response(prompt, temperature=0.7, max_tokens=200):
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

    model.config.pad_token_id = tokenizer.eos_token_id  # Avoid warnings
    output_ids = model.generate(
        input_ids,
        max_new_tokens=max_tokens,
        temperature=temperature,
        do_sample=True,
        top_k=10,
        top_p=0.9,
    )
    response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return response

# Interactive chat function
def chat():
    user_profile = load_user_profile()
    chat_history = load_chat_history()

    # User profile setup if not already done
    if not user_profile:
        print("Let's set up your profile.")
        user_profile["name"] = input("What's your name? ")
        user_profile["age"] = input("How old are you? ")
        user_profile["gender"] = input("What's your gender? ")
        user_profile["country"] = input("Which country are you from? ")
        save_user_profile(user_profile)
        print("Profile saved!")

    print(f"Welcome back, {user_profile['name']}!")
    print("Chatbot is ready! Type 'exit' to quit.")

    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Goodbye! See you next time!")
            save_chat_history(chat_history)  # Save chat history before exiting
            break
        else:
            # Combine chat history with user input for context
            chat_context = "\n".join(chat_history + [f"{user_profile['name']}: {user_input}"])
            response = generate_response(chat_context)
            print(f"Chatbot: {response}")
            chat_history.append(f"{user_profile['name']}: {user_input}")
            chat_history.append(f"Chatbot: {response}")


```

```python
# Start the chatbot
chat()
```
